## **데이터 분석**

```python
import yfinance as yf

# 넷플릭스와 금 펀드
columns = ['NFLX', 'GLD']

# 컬럼에 담긴 종목 데이터를 다운로드 후 시계열 데이터 프레임으로 생성
# 2018년부터 현재 시점까지 , 소수점 4자리까지 표기
f_df = yf.download(columns, start='2018-01-01')['Adj Close'].round(4)
f_df
```

```python
# 데이터 정보
f_df.info()
```

```python
# 데이터 복사
pre_f_df = f_df.copy()
```

```python
# 달러를 원화로 변경하는 함수
def krw_transform(usd):
    usd = usd * 1380
    return usd
```

```python
# 데이터의 달러를 원화로 변경
pre_f_df.GLD = pre_f_df.GLD.apply(lambda x: krw_transform(x))
pre_f_df
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 한글 폰트 설정
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
import matplotlib.pyplot as plt

# 2018 - 2024 넷플릭스와 금 펀드의 주가 시각화
pre_f_df.plot(figsize=(15, 5), subplots=True)
plt.show()
```

- 시대의 발전에 따라 2018년 이후 Netflix는 전체적으로 상승세를 보이고 있으나, 2018년 말, 2020년 초, 2022년 중반 하락한 모양을 보여준다.
- 2018년 말 코로나 19의 영향으로 경제가 하락함에 따라 하락세를 보였을 것이라 추정
- 2020년 초, 멤버십 요금제를 발표하면서 고객의 반발이 있었음
- 2021년 후반, 오징어게임이 에미상에서 미국 드라마로 인정받음과 동시에 6관왕을 차지 국내외로 넷플릭스가 흥행 (상승요인으로 추정)
- 2022년 중반, 코로나 19의 종식에 따라 많은 사용자들이 외부 활동을 시작하면서 하락했을 것이라고 추정

```python
# 차분 후 데이터 시각화
pre_f_df.diff().plot(figsize=(15, 5), subplots=True)
plt.show()
```

- 차분 후 데이터의 변동을 확인해보았을 때, Netflix는 변동의 폭이 넓은 구간이 많아 분산이 일정하지 않다.
- Netflix는 GLD에 비해 안정성이 떨어져보인다고 판단한다.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(7, 5))

# 변화율 시각화
pre_f_df.pct_change().mean().plot(kind='bar', edgecolor='black', ax=ax[0])
ax[0].set_title('변화율')

# 변동률 시각화
pre_f_df.pct_change().std().plot(kind='bar', edgecolor='black', ax=ax[1])
ax[1].set_title('변동률')

plt.tight_layout()
plt.show()
```

- Netflix가 GLD보다 변화율과 변동률이 상대적으로 높다.
- 이는 Netflix의 수익률의 변동이 큼을 의미하고, 장기적으로 안정성이 떨어진다고 볼 수 있다.

```python
import numpy as np

# 수익률 계산
# 각 날짜의 주가를 전날 주가로 나눈 후 로그를 취해 수익률 계산
# 로그를 취해 백분율로 변환
rate_f_df = np.log(pre_f_df / pre_f_df.shift(1))
rate_f_df
```

```python
# 변동률 계산하기
# 변동률 = (오늘 종가 - 어제 종가) / 어제 종가
pct_f_df = pre_f_df.pct_change()
pct_f_df
```

```python
import matplotlib.pyplot as plt

# 그래프 크기 설정
fig, ax = plt.subplots(2, 1, figsize=(10, 9))

# 특정 데이터 수익률 시각화
# lw = line widht = 선 두께
rate_f_df[['NFLX', 'GLD']].plot(lw=0.5, ax=ax[0])
ax[0].set_title('수익률')

# 특정 데이터 수익률 시각화
# lw = line widht = 선 두께
pct_f_df[['NFLX', 'GLD']].plot(lw=0.5, ax=ax[1])
ax[1].set_title('변동률')

# 레이아웃 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```

```python
# 연율화
# 연간 영업일을 252일로 계산
rate_f_df.mean() * 252
```

- GLD의 연간수익률은 8.39%, Netflix의 연간수익률은 18.75%이다.  
- Netflix가 연간 평균적으로 약 2배정도 높은 수익률을 가지고 있지만 큰 변동성과 높은 변화율을 가지고 있으므로  
  안정적인 수익을 선호한다면 GLD, 높은 수익을 추구한다면 Netflix를 선호할 것이라고 추측한다.  

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 다중 공산성 평가 지표 VIF 함수 선언
def get_vif(features):
    vif = pd.DataFrame()
    vif['vif_score'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
    vif['feature'] = features.columns
    return vif
```

```python
# NaN 값 제거 후 VIF 확인
rate_f_df = rate_f_df.dropna()
get_vif(rate_f_df)
```

- Netflix와 GLD는 매우 낮은 상관관계를 보인다. 두 종목은 서로 독립적인 성향을 띄고 있다.

```python
# 데이터 비교의 편의성을 위해 log변환을 하여 분포 조정
# log변환을 통해 바뀐 분포 시각화
fig, ax = plt.subplots(2, 2, figsize=(10, 9))

f_df.NFLX.hist(ax=ax[0][0], bins=50)
f_df.GLD.hist(ax=ax[0][1], bins=50)

rate_f_df.NFLX.hist(ax=ax[1][0], bins=50)
rate_f_df.GLD.hist(ax=ax[1][1], bins=50)

plt.show
```

- 데이터를 로그 변환한 그래프를 보았을 때, 정규 분포에 가까워진 것을 볼 수 있다.  
  이는 일정한 척도로 비교할 수 있는 편의성을 취할 수 있다.  

```python
import matplotlib.pyplot as plt

# 일간 수익률
# cumsum = 각 원소들의 누적합의 과정
# exp: 지수 (로그 후 지수 하면 원래 값을 얻을 수 있음)
rate_f_df.cumsum().apply(np.exp).plot(figsize=(12, 6))
plt.show()
```

```python
# 월간 수익률
# resample: 단위 리샘플링, 1m = 월 단위로 리샘플링
# last: 리샘플링된 그룹에서 가장 마지막 값을 선택 (월간은 가장 마지막 값이 필요)
rate_f_df.cumsum().apply(np.exp).resample('1m').last().plot(figsize=(12, 6))
plt.show()
```

- 일간, 연간 그래프를 확인했을 때 GLD는 수익률이 낮지만 안정적이며, Netflix는 전체적인 수치는 높지만 수익률의 변동이 크다는 것을 확인할 수 있다.

```python
# 넷플릭스의 주가 데이터프레임으로 생성
# Nan 값 제거
n_df = pre_f_df[['NFLX']].dropna()
n_df
```

```python
# 계산할 기간을 20일로 지정
window = 20

# 이동평균 후 최솟값 계산
n_df['min'] = n_df['NFLX'].rolling(window=window).min()
# 이동평균 후 평균값 계산
n_df['mean'] = n_df['NFLX'].rolling(window=window).mean()
# 이동평균 후 최댓값 계산
n_df['max'] = n_df['NFLX'].rolling(window=window).max()

# Nan 값 제거
n_df.dropna()
```

```python
import matplotlib.pyplot as plt

# 최솟값, 평균값, 최댓값의 이동평균 시각화
# 마지막 252일의 데이터 시각화(1년 치)
# 최솟값, 최댓값 초록색 점선으로, 평균값은 빨간색으로 표기
ax = n_df[['min', 'mean', 'max']].iloc[-252:].plot(figsize=(12, 6), style=['g--', 'r--', 'g--'], lw=0.8)
# 마지막 252일의 원본 데이터도 함께 표기
n_df['NFLX'].iloc[-252:].plot(ax=ax)

plt.title("NFLX 20-Day Moving Average Price Movement")
plt.show()
```

- 20일을 기준으로한 이동평균을 확인했을 때, 주가의 상승과 하락의 폭이 넓지만 2023년 11월 이후 전체적으로 상승하는 추세를 보여준다.

```python
# SMA(Simple Moving Average): 일정 기간동안의 가격의 평균을 나타내는 보조지표
# 1달 영업일을 21일로 가정, 1년 영업일을 252일로 가정

n_df['NA1'] = n_df['NFLX'].rolling(window=21).mean() #short-term (단기)
n_df['NA2'] = n_df['NFLX'].rolling(window=252).mean() #long-term (장기)
n_df[['NFLX', 'NA1', 'NA2']].tail()
```

```python
n_df[['NFLX', 'NA1', 'NA2']].plot(figsize=(12, 6))
plt.title("NFLX Simple Moving Average - Long & Short-term")

plt.show()
# 단기가 장기에 비해 내려가면 적극 매도를 해야 하고
# 단기가 장기에 비해 올라가면 골든 크로스로 매수해야 한다.
```

```python
# NaN 값 제거
n_df.dropna(inplace=True)

# 거래 신호 결정 기준 데이터 후 새로운 컬럼으로 추가
# SMA1이 SMA2 보다 크면 1, 작으면 -1
n_df['positions'] = np.where(n_df['NA1'] > n_df['NA2'], 1, -1)  # 1: buy , -1: sell /

# 주식 가격, 이동평균, 거래 신호 데이터 시각화
# secondary_y: 보조 y축 지정 (오른쪽 표기)
ax = n_df[['NFLX', 'NA1', 'NA2', 'positions']].plot(figsize=(15, 6), secondary_y='positions')
# 범례를 조정하여 그래프 우측 상단에 표시
ax.get_legend().set_bbox_to_anchor((-0.05, 1))

plt.title("NFLX Trading Window based on Technical Analysis")
plt.show()
```

- 전체적으로 골든크로스와 데드크로스의 범위가 넓다. 이는 이동평균선의 교차가 일어나는 기간이 길다는 것을 의미하며 변동성이 크다고 볼 수 있다.  
  Netflix는 시장경쟁이 치열하며 주요 수익원이 컨텐츠와 구독료이기 때문에 트렌드에 민감하다.  
  때문에 시장 트렌드에 따라 상승세와 하락세를 판단하며 매수하는 것이 좋다.   

## **머신러닝**

```python
import yfinance as yf
import pandas as pd

# 넷플릭스
columns = ['NFLX']

# 컬럼에 담긴 종목 데이터를 다운로드 후 시계열 데이터 프레임으로 생성
# 2018년부터 현재 시점까지 , 소수점 4자리까지 표기
a_df = yf.download(columns, start='2018-01-01')['Adj Close'].round(4)
# 시리즈를 데이터 프레임으로 변환
a_df = pd.DataFrame(a_df)
a_df
```

```python
# 데이터 정보
a_df.info()
```

```python
# 컬럼명 변경
pre_a_df = a_df.rename(columns={'Adj Close': 'NFLX'})
pre_a_df
```

```python
# 넷플릭스의 최근 1년간 데이터를 데이터 프레임에 담기
nf_df = pre_a_df[['NFLX']].iloc[-252:]
nf_df
```

```python
# 데이터의 80퍼센트를 train, 나머지를 test데이터로 분리
y_train = nf_df.NFLX[:int(0.8 * len(nf_df))]
y_test = nf_df.NFLX[int(0.8 * len(nf_df)):]
```

```python
import matplotlib.pyplot as plt

# train, test데이터 시각화
y_train.plot()
y_test.plot()

plt.show()
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# 그래프 사이즈 지정
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# 정상성 없는 데이터 acf, 차분된 데이터 pacf 계산 및 그래프 생성
# 해당 그래프를 중점으로 확인하기
plot_acf(nf_df, lags=20, ax=ax[0])
plot_pacf(nf_df.diff().dropna(), lags=20, ax=ax[1])


plt.show()
```

```python
from pmdarima.arima import ndiffs
# KPSS(Kwaiatkowski-Phillips-Schmidt-Shin)
# 차분을 진행하는 것이 필요할 지 결정하기 위해 사용하는 한 가지 검정 방법
# 영가설(귀무가설)을 "데이터에 정상성이 나타난다."로 설정한 뒤
# 영가설이 거짓이라는 증거를 찾는 알고리즘이다.

# alpha: 유의수준(기각 영가설 설정, 주로 0.05로 설정)
# test: 단위근 검정 방법 설정
# max_d: 최대차분 횟수 설정
kpss_diffs = ndiffs(y_train, alpha=0.05, test='kpss', max_d=6)
adf_diffs = ndiffs(y_train, alpha=0.05, test='adf', max_d=6)
pp_diffs = ndiffs(y_train, alpha=0.05, test='pp', max_d=6)

n_diffs = max(kpss_diffs, adf_diffs, pp_diffs)

# 최적 차분 횟수
print(f'd = {n_diffs}')
```

```python
import pmdarima as pm

model = pm.auto_arima(y=y_train, 
                      d=1, 
                      start_p=0, max_p=10, 
                      start_q=0, max_q=10, 
                      m=1, seasonal=False, 
                      stepwise=True, 
                      trace=True)
```

```python
model.fit(y_train)
```

```python
# Prob(Q), 융-박스 검정 통계량
# 영가설: 잔차가 백색잡음 시계열을 따른다.
# 0.05 이상: 서로 독립이고 동일한 분포를 따른다.

# Prob(H), 이분산성 검정 통계량
# 영가설: 잔차가 이분산성을 띠지 않는다.
# 0.05 이상: 잔차의 분산이 일정하다.

# Prob(JB), 자크-베라 검정 통계량
# 영가설: 잔차가 정규성을 따른다.
# 0.05 이상: 일정한 평균과 분산을 따른다.

# Skew(쏠린 정도, 왜도)
# 0에 가까워야 한다.

# Kurtosis(뾰족한 정도, 첨도)
# 3에 가까워야 한다.

print(model.summary())
```

```python
import matplotlib.pyplot as plt

model.plot_diagnostics(figsize=(16, 8))
plt.show()
```

```python
# 예측
prediction = model.predict(n_periods=len(y_test))
prediction
```

```python
prediction, conf_int = model.predict(n_periods=1, return_conf_int=True)
print(conf_int)
print(prediction)
```

```python
prediction.tolist()[0]
```

```python
# 예측 함수 선언
def predict_one_step():
    prediction = model.predict(n_periods=1)
    return prediction.tolist()[0]
```

```python
p_list = []

for data in y_test:
    p = predict_one_step()
    p_list.append(p)

    model.update(data)
```

```python
y_predict_df = pd.DataFrame({"test": y_test, "pred": p_list})
y_predict_df
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

plt.plot(y_train.iloc[-50:], label='Train')
plt.plot(y_test.iloc[-50:], label='Test')
plt.plot(y_predict_df.pred, label='Prediction')
plt.legend()
plt.show()
```

```python
import numpy as np

def MAPE(y_test, y_pred):
    return np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print(f'MAPE (%): {MAPE(y_test, p_list):.4f}')
```

- 1.4269% 정도의 오차가 있다.

## **딥러닝**

```python
import yfinance as yf
import pandas as pd

# 넷플릭스
columns = ['NFLX']

# 컬럼에 담긴 종목 데이터를 다운로드 후 시계열 데이터 프레임으로 생성
# 2018년부터 현재 시점까지 , 소수점 4자리까지 표기
a_df = yf.download(columns, start='2018-01-01')['Adj Close'].round(4)
# 시리즈를 데이터 프레임으로 변환
a_df = pd.DataFrame(a_df)
# 인덱스를 연도로 변경
a_df.reset_index(inplace=True)

a_df
```

```python
# 데이터 정보
a_df.info()
```

```python
# Prophet사용을 위해 날짜와 타겟을 ds와 y로 이름 변경
pre_a_df = a_df.rename(columns={'Date': 'ds', 'Adj Close': 'y'})
pre_a_df
```

```python
# 특정 구간 데이터만 추출
pre_a_df = pre_a_df.iloc[-504:].reset_index(drop=True)
pre_a_df
```

```python
from prophet import Prophet

model = Prophet().fit(pre_a_df)
```

```python
future = model.make_future_dataframe(periods=365)

forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 6))

plt.plot(pre_a_df[['y']], label='Train')
plt.plot(forecast[['yhat']], label='Prediction')
plt.legend()
plt.show()
```

```python
model.plot(forecast, figsize=(15, 8), xlabel='year-month', ylabel='price')
plt.show()
```

#### Prophet 하이퍼 파라미터 튜닝

```python
import yfinance as yf
import pandas as pd

# 넷플릭스
columns = ['NFLX']

# 컬럼에 담긴 종목 데이터를 다운로드 후 시계열 데이터 프레임으로 생성
# 2018년부터 현재 시점까지 , 소수점 4자리까지 표기
a_df = yf.download(columns, start='2018-01-01')['Adj Close'].round(4)
# 시리즈를 데이터 프레임으로 변환
a_df = pd.DataFrame(a_df)
# 인덱스를 연도로 변경
a_df.reset_index(inplace=True)

a_df
```

```python
# Prophet사용을 위해 날짜와 타겟을 ds와 y로 이름 변경
pre_a_df = a_df.rename(columns={'Date': 'ds', 'Adj Close': 'y'})
pre_a_df
```

```python
# 특정 구간 데이터만 추출
nf_df = pre_a_df.iloc[-504:].reset_index(drop=True)
nf_df
```

```python
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics
# 모든 경우의 수를 조합해주는 기능
import itertools

# 파라미터 지정
# changepoint_prior_scale: trend(트렌드, 추세)의 크기를 반영하는 정도이다. 0.05가 default
# seasonality_prior_scale: 계절성을 반영하는 단위이다.
# seasonality_mode: 계절성으로 나타나는 효과를 더해나갈지(additive),곱해나갈지 정한다.(multiplicative)
# multiplicative = 예측시 변화가 잘 보이지 않을때 사용. 시간의 흐름에 따라 변화가 커지게 해줌
search_space = {
    'changepoint_prior_scale': [0.05, 0.1, 0.5, 1.0, 5.0, 10.0],
    'seasonality_prior_scale': [0.05, 0.1, 1.0, 10.0],
    'seasonality_mode': ['additive', 'multiplicative']
}

# search_space의 values를 언패킹한 후에 컴프리핸션을 사용하여 튜플 형식의 dict로 만들어준다.
# itertools.product(): 각 요소들의 모든 경우의 수 조합으로 생성
param_combinded = [dict(zip(search_space.keys(), v)) for v in itertools.product(*search_space.values())]

# 전체 데이터 80%를 train, 20%를 test 개수로 저장
train_len = int(len(nf_df) * 0.8)
test_len = int(len(nf_df) * 0.2)

# train, test 사이즈 지정
train_size = f'{train_len} days'
test_size = f'{test_len} days'

# train, test 데이터 분리
train_df = nf_df.iloc[: train_len]
test_df = nf_df.iloc[train_len: ]


# 딕셔너리 개체로 가져올 수 있게 만들어준 후에 언패킹
# model에 담아준다음 fit
mapes = []
for param in param_combinded:
    model = Prophet(**param)
    model.fit(train_df)
    # 교차검증
    # initial :처음 기간
    # period :다음부터 처리될 기간 (임의로 지정)
    # horizen : 배치사이즈
    # threads :메모리 사용량은 낮지만 CPU 바운드 작업에는 효과적이지 않을 수 있다.
    # dask :대규모의 데이터를 처리하는데 효과적이다.
    # processes :각각의 작업을 별도의 프로세스로 실행하기때문에 CPU 바운드 작업에 효과적이지만, 메모리 사용량이 높을 수 있다.
    cv_df = cross_validation(model, initial=train_size, period='20 days', horizon=test_size, parallel='processes')
    # MA(1)
    df_p = performance_metrics(cv_df, rolling_window=1)
    # 각 교차검증 중 mape의 제일 낮은 값 = loss값 가져오기
    mapes.append(df_p['mape'].values[0])
    
# 결과로 데이터프레임 생성
tuning_result = pd.DataFrame(param_combinded)
tuning_result['mape'] = mapes
```

```python
# 결과
tuning_result.sort_values(by='mape')
```

```python
# loss 값이 제일 낮은 파라미터 값 가져와서 최적의 모델로 담기 
model = Prophet(changepoint_prior_scale=10.0, 
                seasonality_prior_scale=10.0, 
                seasonality_mode='additive')

# 훈련
model.fit(nf_df)

# Prophet 모델을 사용하여 미래의 예측값을 생성한다.
# make_future_dataframe: 미래의 일정한 기간에 해당하는 날짜를 포함하는 DataFrame 생성
future = model.make_future_dataframe(periods=365)

# 예측
# 미래의 날짜 정보가 포함된 future DataFrame 사용
forecast = model.predict(future)
# 예측 결과 중 특정 컬럼 추출하여 출력
# ds: 날짜, yhat: 해당 날짜의 예측값, yhat_lower: 예측값 하한, yhat_upper: 예측값 상한
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```python
# 데이터 프레임 복제
pre_nf_df = nf_df.copy()
# 특정 컬럼을 인덱스로 지정
pre_nf_df.set_index('ds', inplace=True)

# 데이터 프레임 복제
forecast_df = forecast.copy()
# 특정 컬럼을 인덱스로 지정
forecast_df = forecast_df.set_index('ds')

pre_nf_df.index = pd.to_datetime(pre_nf_df.index)
forecast_df.index = pd.to_datetime(forecast_df.index)
```

```python
# 그래프 시각화
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

# 훈련 데이터
plt.plot(pre_nf_df[['y']], label='Train')
# 예측 데이터
plt.plot(forecast_df[['yhat']], label='Prediction')

plt.legend()
plt.show()
```

```python
# 예측 결과 데이터 프레임의 정보로 시각화
model.plot(forecast, figsize=(15, 8), xlabel='year-month', ylabel='price')
plt.show()
```

```python
# Prophet 모델의 구성 요소를 시각화
model.plot_components(forecast, figsize=(20, 20))
plt.show()
```
