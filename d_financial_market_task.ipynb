```python
import pandas as pd
import json

# 파일 불러오기
with open('./datasets/bitcoin_2010_2024.json') as f:
    json_data = json.load(f)

# 출력하기
# json_data

# 타입 확인하기
# type(json_data)

# 필요한 특정 데이터 list를 가져와 데이터 프레임 만들기
bitcoin_df = pd.DataFrame(json_data['market-price'])
bitcoin_df
```

```python
from datetime import datetime

# 날짜 형변환 함수 선언
def changeDate(milis):
    # 초로 먼저 바꾸기
    timestamp = milis / 1000
    # datetime 사용해서 초를 연월일로 변환
    converted_time = datetime.fromtimestamp(timestamp).strftime('%Y-%m-%d')
    return converted_time
```

```python
# 날짜 형변환하기
bitcoin_df.x = bitcoin_df.x.apply(changeDate)
bitcoin_df
```

```python
# 날짜 피처를 인덱스로 설정하기
bitcoin_df.set_index('x', inplace=True)
bitcoin_df
```

```python
# 데이터 타입 확인하기
# 현재 타입은 Index. 시계열 데이터로 바꿔줘야 한다.
bitcoin_df.info()
```

```python
# 데이터 타입을 DatetimeIndex로 변환하기
bitcoin_df.index = pd.to_datetime(bitcoin_df.index)
bitcoin_df.info()
```

```python
# train 데이터 80%, test 데이터 20% 나누기
y_train = bitcoin_df['y'][:int(0.8 * len(bitcoin_df))]
y_test = bitcoin_df['y'][int(0.8 * len(bitcoin_df)):]
```

```python
# test와 예측값을 비교하는 그래프 시각화하기
y_train.plot()
y_test.plot()
```

```python
from pmdarima.arima import ndiffs
# KPSS(Kwaiatkowski-Phillips-Schmidt-Shin)
# 차분을 진행하는 것이 필요할지 결정하기 위해 사용하는 한가지 검정 방법
# 영가설(귀무가설)을 "데이터에 정상성이 나타난다" 로 설정한 뒤
# 영가설이 거짓이라는 증거를 찾는 알고리즘이다.

# alpha: 강도
# max_d: 임의로 주되 너무 높으면 패턴을 잃어버랄 수 있다
kpss_diffs = ndiffs(y_train, alpha=0.05, test='kpss', max_d=6)
adf_diffs = ndiffs(y_train, alpha=0.05, test='adf', max_d=6)
pp_diffs = ndiffs(y_train, alpha=0.05, test='pp', max_d=6)

n_diffs = max(kpss_diffs, adf_diffs, pp_diffs)

print(f'd = {n_diffs}')
```

```python
import pmdarima as pm

# d: 차수 설정
# m: 디폴트 값은 1. 계절성이 없을때 사용하며, seasonal= False
# seasonal: 계절성이 있을 경우 True, m에 n을 입력한다.
# stepwise : d > p > q 순서대로 최적의 결과를 찾아준다.
# trace : True로 주면 적합할때마다 출력이된다. 과정을 확인하고싶을때 사용.

model = pm.auto_arima(y=y_train, 
                      d=1, 
                      start_p=0, max_p=10, 
                      start_q=0, max_q=10, 
                      m=1, seasonal=False, 
                      stepwise=True, 
                      trace=True)
```

```python
model.fit(y_train)
```

```python
print(model.summary())
```

```python
# Prob(Q), 융-박스 검정 통계량
# 0.99 으로 백색잡음 시계열에 가깝다. 서로 독립적이며 동일한 분포를 따르고 있다.

# Prob(H), 이분산성 검정 통계량
# 0.00 으로 분산이 일정하지 않다고 판단한다. 이는 하이리스크, 하이리턴의 상태가 존재하며 오차가 크게 나는 부분이 있을 것이라고 예측할 수 있다.

# Prob(JB), 자크-베라 검정 통계량
# 0.00 으로 평균과 분산이 일정하지 않다고 판단한다. (0.05이상 일정하다고 판단)

# Skew(쏠린 정도, 왜도)
# -0.33 으로 0에 가깝지만 왼쪽으로 살짝 치우쳐있다.

# Kurtosis(뾰족한 정도, 첨도)
# 9.44 으로 분포가 첨도가 높다. 뾰족하다.

# 해당 모델은 `융-박스 검정 통계량`을 제외한 나머지 수치가 적절하지 않으므로 지속적으로 동일한 수익률을 나타내기는 어렵습니다.
# 하지만 왜도와 첨도를 봤을때 0에 가까우며 뾰족한 형태의 분포를 띄고 있다고 판단할 수 있습니다.
```

```python
# summary 결과를 시각화
import matplotlib.pyplot as plt

model.plot_diagnostics(figsize=(16,8))
plt.show()
```

```python
prediction = model.predict(n_periods=len(y_test))
prediction
```

```python
prediction, conf_int = model.predict(n_periods=1, return_conf_int=True)
print(conf_int)
print(prediction)
```

```python
# 예측 할 때마다 (n_periods=1) 그 결과를 리스트로 리턴해주는 함수 생성
# 결과: 신뢰구간의 평균값
# 예측 할 때마다 핏을 한 다음 다음을 예측하기 위함
def predict_one_step():
    prediction = model.predict(n_periods=1)
    return prediction.tolist()[0]
```

```python
p_list = []

# 예측 할 때마다 예측값을 리스트에 담아준다.
for data in y_test:
    p = predict_one_step()
    p_list.append(p)

    model.update(data)
```

```python
# 테스트 데이터와 예측 데이터로 데이터프레임 생성
y_predict_df = pd.DataFrame({"test": y_test, "pred": p_list})
y_predict_df
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

# 실제 데이터와 얼만큼 차이가 나는지 그래프로 시각화
plt.plot(y_train, label='Train')
plt.plot(y_test, label='Test')
plt.plot(y_predict_df.pred, label='Prediction')
plt.legend()
plt.show()
```

```python
import numpy as np

# MAPE로 평가하는 함수 생성
def MAPE(y_test, y_pred):
    return np.mean(np.abs((y_test - y_pred) / y_test)) * 100

# 오차 퍼센트를 출력
print(f'MAPE (%): {MAPE(y_test, p_list):.4f}')
```
